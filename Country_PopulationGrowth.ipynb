```python
import csv
import re
import matplotlib.pyplot as plt

#READ_CSV
def read_csv(path):
  with open(path,'r') as csvfile:
    reader = csv.reader(csvfile, delimiter = ',')
    header = next(reader)
    data = []
    for row in reader:
      iterable = zip(header, row)
      country_dict = {key:value for key,value in iterable}
      data.append(country_dict)
    return data
```

```python
#UTILS
def get_population(country_dict):
  newKeys = []
  newValues = []

  r = re.compile("^[0-9]*")
  
  for key in country_dict:
    x = r.match(key)
    if x.group() != '':
      newKeys.append(x.group())
      newValues.append(country_dict[key])

  new_dict = dict(zip(newKeys,newValues))
  return new_dict


def popul_by_country(data, country):
  result = list(filter(lambda item : item['Country/Territory'] == country,data))
  return result
```

```python
#CHARTS
def generate_bar_chart(labels,values):
  fig, ax = plt.subplots()
  ax.bar(labels,values)
  plt.show()

def generate_pie_chart(labels,values):
  fig, ax = plt.subplots()
  ax.pie(values, labels=labels)
  ax.axis('equals')
  plt.show()
```

```python
#MAIN
def run():
  data = read_csv('/content/data.csv')
  print('Data: ', data)

  country = input('Type a country: ')

  result = popul_by_country(data,country)

  result_dict = result[0]
  print('\n',result_dict)

  new_dict = get_population(result_dict)
  print('\n',new_dict)
  
  generate_bar_chart(new_dict.keys(),new_dict.values())
  

if __name__ == '__main__':
  run()
```
